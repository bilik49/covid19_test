#### 4. Выведите таблицу, которая позволит просмотреть первую дату поставки и общее количество поставленной вакцины в рамках Марки вакцины и страны.

```python
# ВАШ КОД ЗДЕСЬ
df_grouped = df.groupby(['Марка вакцины', 'Страна'], as_index = False).agg({'Дата поставки/новости': 'min', 'Вакцина (кол-во)': 'sum'})
df_grouped.rename(columns = {'Дата поставки/новости': 'Первая дата поставки'}, inplace = True)
df_grouped
```

#### 6. Достать домены источников 1, 2, 3
Для столбцов Источников создать столбцы Доменов, где будет извлечен только домен (источник: https://sputnik-abkhazia.ru/Abkhazia/20200320/..., домен: sputnik-abkhazia.ru). В исходном датасете 3 столбца источника и соответственно должно быть создана 3 столбца с доменами.

```python
# ВАШ КОД ЗДЕСЬ
from urllib.parse import urlparse
df['Домен 1'] = df['Источник 1'].apply(lambda x: urlparse(x).netloc if pd.notnull(x) else np.nan)
df['Домен 2'] = df['Источник 2'].apply(lambda x: urlparse(x).netloc if pd.notnull(x) else np.nan)
df['Домен 3'] = df['Источник 3'].apply(lambda x: urlparse(x).netloc if pd.notnull(x) else np.nan)
df
```
